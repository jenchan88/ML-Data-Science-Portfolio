```python
X_train, X_dev, y_train, y_dev = train_test_split(X, y, test_size=0.1, random_state=42, stratify=y)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_dev)

ldaModel = LinearDiscriminantAnalysis()
ldaModel.fit(X_train, y_train)

train_pred = ldaModel.predict(X_train)
y_pred = ldaModel.predict(X_dev)


# Training set metrics
train_accuracy = accuracy_score(y_train, train_pred)
train_precision = precision_score(y_train, train_pred, average='weighted')
train_recall = recall_score(y_train, train_pred, average='weighted')
train_f1 = f1_score(y_train, train_pred, average='weighted')
train_confusion_matrix = confusion_matrix(y_train, train_pred)

# Test set metrics
test_accuracy = accuracy_score(y_dev, y_pred)
test_precision = precision_score(y_dev, y_pred, average='weighted')
test_recall = recall_score(y_dev, y_pred, average='weighted')
test_f1 = f1_score(y_dev, y_pred, average='weighted')
test_confusion_matrix = confusion_matrix(y_dev, y_pred)

# Print results
print("Training Metrics:")
print(f"Accuracy: {train_accuracy:.4f}")
print(f"Precision: {train_precision:.4f}")
print(f"Recall: {train_recall:.4f}")
print(f"F1 Score: {train_f1:.4f}")
print(f"Confusion Matrix:\n{train_confusion_matrix}\n")

print("Testing Metrics:")
print(f"Accuracy: {test_accuracy:.4f}")
print(f"Precision: {test_precision:.4f}")
print(f"Recall: {test_recall:.4f}")
print(f"F1 Score: {test_f1:.4f}")
print(f"Confusion Matrix:\n{test_confusion_matrix}\n")

print("Classification Report (Test Set):\n", classification_report(y_test, test_pred))
```
