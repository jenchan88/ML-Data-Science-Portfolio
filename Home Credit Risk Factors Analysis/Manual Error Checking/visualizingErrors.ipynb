```python
import pandas as pd
from google.colab import drive
import matplotlib.pyplot as plt
import seaborn as sns
drive.mount('/content/drive')
```

```python
lr_errors = pd.read_excel("/content/drive/MyDrive/data403project2/lr_errors.xlsx")
```

```python
# 1. Target Distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=lr_errors, x='TARGET', palette="viridis")
plt.title("Distribution of Loan Targets")
plt.xlabel("Target (0 = Repaid, 1 = Defaulted)")
plt.ylabel("Count")
plt.show()

```

```python
# 2. Income vs. Credit Amount by Target
plt.figure(figsize=(10, 8))
sns.scatterplot(data=lr_errors, x='AMT_INCOME_TOTAL', y='AMT_CREDIT', hue='TARGET', alpha=0.6)
plt.title("Income vs. Credit Amount by Target")
plt.xlabel("Total Income")
plt.ylabel("Credit Amount")
plt.legend(title="Target")
plt.show()
```

```python
# 3. Loan Amount by Contract Type
plt.figure(figsize=(10, 8))
sns.boxplot(data=lr_errors, x='NAME_CONTRACT_TYPE', y='AMT_CREDIT', hue='TARGET', palette="pastel")
plt.title("Loan Amount by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Credit Amount")
plt.yscale("log")
plt.show()
```

```python
# 4. Correlation Heatmap
numerical_columns = ['AMT_INCOME_TOTAL', 'AMT_CREDIT', 'AMT_ANNUITY', 'DAYS_BIRTH', 'DAYS_EMPLOYED', 'DAYS_REGISTRATION']
plt.figure(figsize=(12, 10))
sns.heatmap(lr_errors[numerical_columns].corr(), annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```
